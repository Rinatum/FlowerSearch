```python
img_sizes = [sam['image'].shape for sam in ds.as_numpy_iterator()]
img_sizes
```

```python
sz = Counter(img_sizes)
str_keys = [ str(key) for key in list(sz.keys()) ]
fig = plt.figure(figsize=(12,8), dpi= 100)
# plt.xticks(fontsize=3, rotation=90)
# plt.bar(str_keys, sz.values())
sns.set(rc={'figure.figsize':(130,50)})
sns.barplot(x = str_keys, y = list(sz.values()) )
```

```python
img_biggest_dim = [max(sam['image'].shape) for sam in ds.as_numpy_iterator()]
img_biggest_dim
```

```python
sbd = Counter(img_biggest_dim)
# print(sbd)
str_keys = [ str(key) for key in list(sbd.keys()) ]
fig = plt.figure(figsize=(12,8), dpi= 200)
# plt.yticks(fontsize=14, rotation=90)
# plt.xticks(fontsize=3, rotation=90)
# plt.bar(str_keys, sbd.values())
sns.set(rc={'figure.figsize':(130,50)})
sns.barplot(x = str_keys, y = list(sbd.values()) )
```

```python
labels = [ sam['label'] for sam in ds.as_numpy_iterator() ]
labels
```
